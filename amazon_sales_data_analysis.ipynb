```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/amazonsales sql.csv")
df.head()
```

##Sales Trend Over Time

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
sales_trend=df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

plt.figure(figsize=(12,6))
sales_trend.plot(marker='o',color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

##Top 10 Best-Selling Products

```python
top_products=df.groupby('Product Name')['Sales'].sum().nlargest(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar',color='g')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45,ha='right')
plt.grid(True)
plt.show()
```

##Sales Distribution by Category

```python
category_sales=df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(12,8))
category_sales.plot(kind='pie',autopct='%1.1f%%',startangle=140,pctdistance=0.85,labeldistance=1.15,wedgeprops={'linewidth':1,'edgecolor':'black'},colormap='viridis')
plt.title('Sales Distribution by Category')
plt.tight_layout()
plt.ylabel('')
plt.show()
```

##Profit Vs.Sales Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Profit'],df['Sales'],alpha=0.7,s=20,color='r')
plt.title('Profit Vs. Sales Scatter Plot')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.xscale('log')
plt.yscale('log')
plt.grid(True,linestyle="--")
plt.show()
```
